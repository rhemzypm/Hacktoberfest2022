```python
import cv2
from matplotlib import pyplot as plt
import numpy as py
from google.colab.patches import cv2_imshow
```

```python
#Nomer 1
image = cv2.imread("cewek.jpg")
gaussian_img = cv2.GaussianBlur(image, (5,5), 0)
cv2_imshow(gaussian_img)
cv2.waitKey(0)

```

```python
#Nomer 2
gaussian_highpass = image - cv2.GaussianBlur(image, (21,21), 3)+127
cv2_imshow(gaussian_highpass)
cv2.waitKey(0)
```

```python
image_burem = cv2.imread("cewek_burem.jpg")
```

```python
#Nomer 3
b,g,r = cv2.split(image_burem)
rgb_img = cv2.merge([r,g,b])
dst = cv2.fastNlMeansDenoisingColored(image_burem,None,40,40,24,40)
b,g,r = cv2.split(dst)
rgb_dst = cv2.merge([r,g,b])
```

```python
#subplots
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original')
ax1.imshow(rgb_img)

ax2.set_title('butterworth notch filter')
ax2.imshow(rgb_dst)
```
